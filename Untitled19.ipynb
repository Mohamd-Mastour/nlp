```python
# Nouveaux exemples de textes
text1 = "The cat sat on the windowsill, enjoying the sunshine."
text2 = "A feline rested by the window, basking in the warm sunlight."
```

```python
# ---------- 1. Cosine Similarity (using TF-IDF Vectorizer) ----------
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([text1, text2])
cos_sim = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
print(f"Cosine Similarity: {cos_sim[0][0]:.4f}")
```

```python
# ---------- 2. Jaccard Similarity ----------
def jaccard_similarity(str1, str2):
    words1 = set(str1.lower().split())
    words2 = set(str2.lower().split())
    intersection = words1.intersection(words2)
    union = words1.union(words2)
    return float(len(intersection)) / len(union)
```

```python
jac_sim = jaccard_similarity(text1, text2)
print(f"Jaccard Similarity: {jac_sim:.4f}")
```

```python
# ---------- 3. Semantic Similarity using Transformers ----------
model = SentenceTransformer('all-MiniLM-L6-v2')  # small and fast
embedding1 = model.encode(text1)
embedding2 = model.encode(text2)
```

```python
semantic_cos_sim = cosine_similarity([embedding1], [embedding2])
print(f"Transformer-based Semantic Similarity: {semantic_cos_sim[0][0]:.4f}")
```
